```python
import pandas
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
import string
```

<h1>Loading in URLs For Each Character</h1>

```python
characters = "https://enjoy-sfv-more.com/en/frames/"
url = "https://enjoy-sfv-more.com/en/frames/sagat"
```

```python
charres = requests.get(characters)
```

```python
charsoup = BeautifulSoup(charres.text, "html.parser")
```

```python
names = charsoup.find_all("li",{"class":"characterLinkItem_item__1F7cn"})
```

```python
unclean = [x.text.lower() for x in names]

cleaned_list = []
for word in unclean:
    cleaned_word = []
    for char in word:
        if char.isalpha() == True:
            cleaned_word.append(char)
    cleaned_list.append("".join(cleaned_word))
```

```python
url_list = []
for x in cleaned_list:
    url_list.append(characters + x)
```

<h1>Making DataFrame for Each Character</h1>

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
table = soup.find_all("table")
loc_header = soup.find("table")
headers = []
for i in loc_header.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
with open('character_names.txt','w') as f:
    for x in cleaned_list:
        f.write(x + "_df" + " = pd.DataFrame(columns = headers)")
        f.write("\n")
f.close()

with open('character_names.txt','a') as f:
    for x in cleaned_list:
        f.write(x + "_df,")
        f.write("\n")
f.close()
```

```python
i = 0 
for x in url_list:
    res = requests.get(x)
    soup = BeautifulSoup(res.text,"html.parser")
    table = soup.find_all("table")

    for x in table:
        value_list = []
        for val in x.find_all('td'):
            values = val.text
            value_list.append(values)
        df_list[i].loc[len(df_list[i].index)] = value_list
    
    moves = []
    for val in soup.find_all("p",{"class":"item_name__3rpyR"}):
        moves.append(val.text)
    df_list[i]["Move_Name"] = moves
    df_list[i].set_index("Move_Name",inplace=True)
    i += 1
```
